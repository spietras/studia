```python
import algorithm
import program
```

```python
start_n = 100
step = 100
iterations = 50
instances = 1
repeats = 1

densities = [0.0, 0.5, 1.0]
```

```python
cumulated_results = []

for density in densities:
    cumulated_results.append(program.measure(start_n, iterations, step, density, instances, repeats, verbosity=2))
```

```python
import statistics

for results, density in zip(cumulated_results, densities):
    points, times = map(list, zip(*results))
    median_point = statistics.median(points)
    median_time = statistics.median(times)

    c = median_time / algorithm.T(median_point, density)

    Ts = list()

    for point, time in results:
        Ts.append(algorithm.T(point, density) * c)

    c3 = median_time / median_point**3

    N3s = list()

    for point, time in results:
        N3s.append(point**3 * c3)

    import matplotlib.pyplot as plt

    plt.plot(points, times, label='$t(n)$')
    plt.plot(points, Ts, label='$cT(n)$')
    plt.plot(points, N3s, label="$cn^3$")

    plt.title("density = {}".format(density))
    plt.xlabel('n')
    plt.legend()
    plt.show()
```

```python
for results, density in zip(cumulated_results, densities):
    program.print_table(results, density)
```
