```python
from protosound.data import FSD50K,FSD50KSpectro
from matplotlib import pyplot as plt
import random
#data_path_name = "C:\\Users\\Arkadiusz SP\\Documents\\data_test"
data_path_name = "C:\\Users\\Arkadiusz SP\\Documents\\Data_ZSN_Final"
random_sample_count = 9




```

```python
a=FSD50K(data_path_name)
spekt= FSD50KSpectro(data_path_name)
random_spect = []
length = a.__len__()
for i in range(random_sample_count):
    index = int(round(random.random()*length,0))
    random_spect.append((spekt.__getitem__(index)[0].transpose(1,2,0),a.items[index][1],a.items[index][0]))
    print(random_spect[i][1])
```

```python
fig = plt.figure(figsize=(20,20))
rows = 3
columns = 3
for i in range(random_sample_count):
   
    fig.add_subplot(rows, columns, i+1)
    plt.imshow(random_spect[i][0])
    #plt.axis('off')
    plt.title("Sound:"+random_spect[i][1]+"\nfile:"+random_spect[i][2], loc='left')
```
